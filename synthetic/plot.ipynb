```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
y = [0.02651515152,0.06742424242,0.08584280303,0.08892045455]
X = ['2','16','32','64']
mod_hi = 5
mod_lo = 4
ylabel = 'Performance gap'
xlabel = 'Number of concepts'
```

```python
fig, ax = plt.subplots()
ax.plot(X, y)  
ax.set_title('{} vs. {}'.format(xlabel, ylabel))
ax.set_xlabel(xlabel, fontsize=15)
ax.set_ylabel(ylabel, fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
fig.savefig('/home/yuncheng/MultiBench/synthetic/concepts_perf{}{}.png'.format(mod_hi, mod_lo), bbox_inches='tight')
```
